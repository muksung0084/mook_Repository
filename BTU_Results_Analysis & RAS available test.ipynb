```python
z=df2
#ë©”ì¸ë„ì–´ ì˜¤í”ˆìƒíƒœ ì§€ì†ì‹œê°„ CHECK
z['Skim dor count'] = np.where((z['FceTemp']>800) & (z['MainDoor_Status']=='Open'), 1, 0)

#MELTING ì‹œìž‘ì‹œì , ëì‹œì , PROGRESS ìµœëŒ€ì .(100%ë„ë‹¬í•˜ëŠ”ì§€ í™•ì¸)
a=z.groupby('BATCHID')['MeltingProgress','Date'].max()
b=z.groupby('BATCHID')['_Timestamp'].min()
c= a.merge(b, left_index=True, right_index=True, how='inner')
c=c.rename(columns={'Date' : 'max_time', '_Timestamp': 'min_time', 'MeltingProgress':'Max_MeltingProgress'})

#ìŠ¤í‚¤ë°ì‹œì  í¬ì¸íŠ¸ (OPEN STATE 4ë¶„ì´ìƒ ìœ ì§€ë˜ëŠ” BATCH) & ê·¸ë•Œì˜ batchë“¤ ì„ ì •
z2 =z.loc[z.groupby('BATCHID')['Skim dor count'].cumsum()==2].groupby('BATCHID')['_Timestamp','MachineName','Alloy','casttype','MainDoor_Status','MeltingProgress','CycleTime','Total_Melting_Time','FceTemp','theo_q_total','theo_q_solid','cold_total','ACTUAL_BTU','month','day'].first()
z2=z2.sort_values(by=['_Timestamp'],inplace=False)
```

```python
z3=z2.merge(c, left_index=True, right_index=True, how='inner').reset_index()
z4=z3[['BATCHID','MachineName','CycleTime','Alloy','casttype','cold_total','min_time','max_time','MeltingProgress','_Timestamp','FceTemp','Max_MeltingProgress','theo_q_total','theo_q_solid','ACTUAL_BTU','month','day' ]]
z4=z4.rename(columns={'_Timestamp':'skimming point(predict)','MeltingProgress':'Skimming_MeltingProgress','Max_MeltingProgress':'Max','ACTUAL_BTU':'Skim_actual_BTU'})

z4['min_time']=pd.to_datetime(z4['min_time'])
z4['max_time']=pd.to_datetime(z4['max_time'])
z4['skimming point(predict)']=pd.to_datetime(z4['skimming point(predict)'])

#ì‹¤ì œ ì–»ì€ dataë¥¼ í† ëŒ€ë¡œí•œ melting cycle (ë-ì‹œìž‘ì‹œì ) ê³¼ skimming point ì‹œì  ë° êµ¬ê°„ êµ¬í•˜ê¸° ( 4ë¶„ ì´ìƒ openì´ ìœ ì§€ë˜ëŠ” ì‹œìž‘ì‹œì )
z4['cycle']=(z4['max_time']-z4['min_time']).dt.seconds/60
import datetime
z4['skimming point(predict)']=z4['skimming point(predict)']-datetime.timedelta(minutes=3)
z4['skim point(min)']=(z4['skimming point(predict)']-z4['min_time']).dt.seconds/60

#ìµœì¢… data ì»¬ëŸ¼ ì •ë¦¬
z4=z4[['min_time','BATCHID','MachineName','Alloy','casttype','theo_q_solid','theo_q_total','cold_total','skimming point(predict)','skim point(min)','FceTemp','Skimming_MeltingProgress','Max','Skim_actual_BTU','month','day' ]]
z44=z4[z4['MachineName'].isin(['Melter4'])].set_index('min_time').loc['2020-05-18':].reset_index().sort_values(by=['min_time'],inplace=False) 
z43=z4[z4['MachineName'].isin(['Melter3'])].set_index('min_time').loc['2020-05-18':].reset_index().sort_values(by=['min_time'],inplace=False) 

mf3_lasttime=mf3[['BATCHID','Date','CycleTime','Requrited_BTU','ACTUAL_BTU']]
mf4_lasttime=mf4[['BATCHID','Date','CycleTime','Requrited_BTU','ACTUAL_BTU']]
z43=z43.merge(mf3_lasttime, on='BATCHID',how='inner')
z44=z44.merge(mf4_lasttime, on='BATCHID',how='inner')

a=z43['Max']-z43['Skimming_MeltingProgress']
b=z44['Max']-z44['Skimming_MeltingProgress']

z43['peak_predicted']=np.where(a>3,2,1)
z44['peak_predicted']=np.where(b>3,2,1)
z43['compliance']=np.where(z43['Skimming_MeltingProgress']< 94,'ok','no')
z44['compliance']=np.where(z44['Skimming_MeltingProgress']< 94,'ok','no')


z43=z43[['BATCHID','MachineName','min_time','month','day','skimming point(predict)','Skimming_MeltingProgress','Max','compliance','peak_predicted','Date','Skim_actual_BTU','theo_q_solid','cold_total','Requrited_BTU','ACTUAL_BTU','theo_q_total','Alloy','casttype','CycleTime']]
z44=z44[['BATCHID','MachineName','min_time','month','day','skimming point(predict)','Skimming_MeltingProgress','Max','compliance','peak_predicted','Date','Skim_actual_BTU','theo_q_solid','cold_total','Requrited_BTU','ACTUAL_BTU','theo_q_total','Alloy','casttype','CycleTime']]
```
